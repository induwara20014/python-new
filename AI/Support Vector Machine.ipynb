```python
import pandas as pd
from pyswip import predicate_t

dataset=pd.read_csv(r'C:\Users\ASUS\Downloads\linear-insperable-dataset.csv').values

print(dataset)
```

```python
data = dataset[:,0:2]
target=dataset[:,2]

print('data :',dataset)
print('target :',target)
```

```python
from matplotlib import pyplot as plt

x=data[:,0]
y=data[:,1]

for i in range(len(data)):

    if(target[i]==0):
        plt.scatter(x[i],y[i],c='r')
    elif(target[i]==1):
        plt.scatter(x[i],y[i],c='g')

plt.xlabel('Feature 1 (x)')
plt.ylabel('Feature 2 (y)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.1)
```

```python
from sklearn.svm import SVC
#from sklearn.neighbors import KNeighborsClassifier

model=SVC(kernel='linear')
#model = KNeighborsClassifier()
model.fit(train_data,train_target)
```

```python
predicate_target=model.predict(test_data)

from sklearn.metrics import accuracy_score

acc=accuracy_score(test_target,predicate_target)
print('Accuracy;',acc)
```
