```python
import pandas as pd

dataset=pd.read_csv(r'C:\Users\ASUS\Downloads\linear-insperable-dataset.csv').values

print(dataset)
```

```python
data = dataset[:,0:2]
target=dataset[:,2]

print('data :',dataset)
print('target :',target)
```

```python
from matplotlib import pyplot as plt

x=data[:,0]
y=data[:,1]

for i in range(len(data)):

    if(target[i]==0):
        plt.scatter(x[i],y[i],c='r')
    elif(target[i]==1):
        plt.scatter(x[i],y[i],c='g')

plt.xlabel('Feature 1 (x)')
plt.ylabel('Feature 2 (y)')
plt.show()
```
