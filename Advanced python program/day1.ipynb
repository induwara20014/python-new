```python
def calAvg(a,b,c):
    avg=(a+b+c)/3.0
    print('avg:',avg)
```

```python
while(True):
    maths,phy,chem=eval(input("Enter Student Marks:"))
    calAvg(maths,phy,chem)

    option=input('Do you need another entry(y/n):')

    if(option=='n' or option=='N'):
        break

print('LOOP ENDS')
```

```python
while(True):

    maths,phy,chem=eval(input("Enter Student Marks:"))

    if (maths<0 or phy<0 or chem<0):

        print('Marks cannot be NEGATIVE!')
        continue

    calAvg(maths,phy,chem)

    option=input('Do you need anther entry(y/n):')

    if(option=='n' or option=='N'):

        break
    print('LOOP ENDS')
```
