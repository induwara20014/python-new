```python
import cv2 #importing opencv module

img = cv2.imread(r"C:\Users\ASUS\OneDrive\Desktop\python new\codes\new-flower.png")
#print(img.shape,img.ndim,type(img))

img[:,100:110]=[0,255,255]

cv2.imshow('LIVE',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
