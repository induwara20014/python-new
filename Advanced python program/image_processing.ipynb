How to load a image

```python
import cv2 #importing opencv module

img = cv2.imread(r"C:\Users\ASUS\OneDrive\Desktop\python new\codes\new-flower.png")
#print(img.shape,img.ndim,type(img))

img[:,100:110]=[0,255,255]

cv2.imshow('LIVE',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.imread(r"C:\Users\ASUS\OneDrive\Desktop\python new\codes\new-flower.png")

img[100:150,100:200]=[0,255,255]

cv2.imshow('IMG',img)
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

How to load a video file

```python
import cv2

source=cv2.VideoCapture(r"C:\Users\ASUS\OneDrive\Desktop\python new\codes\Samples\road_drive.avi")

while(True):
   ret,img=source.read()
   print(ret)

   if(ret==False):
       break

   cv2.imshow('IMG',img)
   cv2.waitKey(1)

cv2.destroyAllWindows()
```
