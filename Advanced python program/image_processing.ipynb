```python
import cv2

source=cv2.VideoCapture(r"C:\Users\ASUS\OneDrive\Desktop\python new\codes\Samples\road_drive.avi")

while(True):
   ret,img=source.read()
   print(ret)

   if(ret==False):
       break

   cv2.imshow('IMG',img)
   key=cv2.waitKey(1)

   if(key==27):
      break
   elif(key==112):
       cv2.waitKey(0)

cv2.destroyAllWindows()
```

Changing Colorspaces

```python
import numpy as np

array=np.random.randint(0,255,(3,5))
print(array)
print('---------------------------------------')
array_new=np.mean(array,axis=1)
print(array_new)
```

```python
def myCvtColor(img):
    img_new=np.mean(img,axis=2)
    img_new=np.array(img_new,dtype=np.uint8)
    print(img.shape,img_new.shape)
    return img_new
```

```python
img=cv2.imread(r"C:\Users\ASUS\OneDrive\Desktop\python new\codes\new-flower.png")
transformed=myCvtColor(img)

cv2.imshow('transformed',transformed)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.imread(r"C:\Users\ASUS\OneDrive\Desktop\python new\codes\new-flower.png")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

cv2.imshow('IMG',img)
cv2.imshow('GRAY',gray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.imread(r"C:\Users\ASUS\OneDrive\Desktop\python new\codes\new-flower.png")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

cv2.imshow('IMG',img)
cv2.imshow('GRAY',gray)
cv2.imshow('HSV',hsv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
