Changing Colorspaces

```python
import numpy as np

array=np.random.randint(0,255,(3,5))
print(array)
print('---------------------------------------')
array_new=np.mean(array,axis=1)
print(array_new)
```

```python
def myCvtColor(img):
    img_new=np.mean(img,axis=2)
    img_new=np.array(img_new,dtype=np.uint8)
    print(img.shape,img_new.shape)
    return img_new
```

```python
img=cv2.imread(r"C:\Users\ASUS\OneDrive\Desktop\python new\codes\new-flower.png")
transformed=myCvtColor(img)

cv2.imshow('transformed',transformed)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

Image Thresholding

```python
img=cv2.imread(r"C:\Users\ASUS\OneDrive\Desktop\python new\codes\new-flower.png")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

ret,thresh1 = cv2.threshold(gray,80,255,cv2.THRESH_BINARY)
ret,thresh_TOZERO = cv2.threshold(gray,80,255,cv2.THRESH_TOZERO)
ret,thresh_TRUNC = cv2.threshold(gray,80,255,cv2.THRESH_TRUNC)

cv2.imshow('IMG',img)
cv2.imshow('gray',gray)
cv2.imshow('thresh_TOZERO',thresh_TOZERO)
cv2.imshow('thresh_TRUNC',thresh_TRUNC)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.imread(r"\Users\ASUS\OneDrive\Desktop\python new\codes\Samples\gray_image.png")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

ret,thresh = cv2.threshold(gray,80,255,cv2.THRESH_BINARY)

cv2.imshow('IMG',img)
cv2.imshow('gray',thresh)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

Color Segmentation

```python
import cv2
import numpy as np

frame=cv2.imread(r"C:\Users\ASUS\OneDrive\Desktop\python new\codes\Samples\colors3.jpg")
hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

lower_color = np.array((96,227,130))
upper_color = np.array((140,255,255))

mask = cv2.inRange(hsv, lower_color, upper_color)

cv2.imshow('mask',mask)
cv2.imshow('hsv',hsv)
cv2.imshow('frame',frame)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
