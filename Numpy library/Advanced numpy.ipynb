```python
import pandas as pd

dataset=pd.read_csv('student-marks.csv')
```

```python
print(dataset)
```

```python
dataset=pd.read_csv('student-marks.csv').values
print(dataset,dataset.ndim,dataset.shape)
```

```python
import numpy as np
avg=np.mean(dataset[:,1:4],axis=1)
print(avg)
```

```python
def findGrade(mark):
    if(mark>=0 and mark<35):
        grade = 'F'
    elif(mark>=35 and mark<45):
        grade = 'S'
    elif(mark>=45 and mark<65):
        grade = 'C'
    elif(mark>=65 and mark<75):
        grade = 'B'
    elif(mark>=75 and mark<=100):
        grade = 'A'
    return grade
```

```python
for i in dataset[:,1]:
    print(findGrade(i))
```

```python
mathsGrades = np.array([findGrade(i) for i in dataset[:,1]])

print(mathsGrades)
mathsGrades=mathsGrades.reshape(-1,1)
print(mathsGrades)
```

```python
phyGrades=np.array([findGrade(i) for i in dataset[:,1]])
phyGrades=phyGrades.reshape(-1,1)
print(phyGrades)
```

```python
chemGrades=np.array([findGrade(i) for i in dataset[:,3]])
chemGrades=chemGrades.reshape(-1,1)
print(chemGrades)
```

```python
avg=avg.reshape(-1,1).reshape(-1,1)
print(avg)
```

```python
print(dataset)
```

```python
dataset=np.append(dataset,mathsGrades,axis=1)
print(dataset)
```

```python
dataset=np.append(dataset,phyGrades,axis=1)
print(dataset)
```

```python
dataset=np.append(dataset,avg,axis=1)
print(dataset)
```
