Indexing & Slicing

```python
array=np.random.randint(0,100,(10,5))
print(array)
```

```python
array1=np.random.randint(0,100,(4,3))
array2=np.random.randint(0,100,(4,3))

array3 = array1 + array2
print('array1:',array1)
print('array2:',array2)
print('array3:',array3)
```

```python
array1=np.random.randint(0,100,(4,3))

array2=array1/10

print('array1:',array1)
print('array2:',array2)
```

```python
array1=np.random.randint(0,100,(4,3))
array2=np.random.randint(0,100,(3,5))

print('array1:',array1)
print('array2:',array2)

array4=np.matmul(array1,array2)

print('array4:',array4)
```
