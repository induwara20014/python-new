```python
value1=eval(input('Enter a value for value1: '))
value2=eval(input('Enter a value for value2: '))

print("addition: ", value1 + value2)
print("sub: ", value1 - value2)
print("multiplication: ", value1 * value2)
print("division: ", value1 / value2)
print("modules: ", value1 % value2)
print("exponent: ", value1 ** value2)
```
